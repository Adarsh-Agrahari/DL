# Import Library

```python
import math
```

# Step Function

```python
def step(x):
  if x>=0:
    return 1
  return 0
```

```python
step(55)
```

```python
step(-20)
```

# Sigmoid Function

```python
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

# Tanh Function

```python
def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

# ReLU Function

```python
def relu(x):
  return max(0,x)
```

```python
relu(-7)
```

```python
relu(1)
```

```python
relu(6)
```

# Leaky ReLU Function

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
